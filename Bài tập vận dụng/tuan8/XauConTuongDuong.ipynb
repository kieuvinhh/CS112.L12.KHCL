# Đề bài:

XÂU CON TƯƠNG ĐƯƠNG (1s, 50M)

Cho xâu s chỉ chứa các ký tự la tinh thường độ dài n không quá 10^5. Ký hiệu s**[**i..j**]** là xâu con các ký tự liên tiếp nhau của s từ vị trí i đến vị trí j. Hai xâu con x = s**[**a..b] và y = s**[**c..d**]** được gọi là cùng một lớp tương đương nếu có cách đổi chỗ các ký tự trong một xâu để nhận được xâu kia. Các vị trí được đánh số từ 1 trở đi.

Ví dụ với s = “abcbacaac”, s**[**1..3**]** tương đương với s**[**4..6**]**, còn s**[**1..3**]** không tương đương với s**[**5..7**]**.

Cho q truy vấn, mỗi truy vấn là 4 số nguyên a, b, c, d xác định các xâu con x và y. Với mỗi truy vấn hãy xác định hai xâu con này có cùng một lớp tương đương hay không và đưa ra câu trả lời tương ứng “YES” hoặc “NO”.

Dữ liệu: Vào từ thiết bị nhập chuẩn:

Dòng đầu tiên chứa xâu s,

Dòng thứ 2 chứa số nguyên q (1 ≤  q  ≤ 10^5),

Mỗi dòng trong q dòng tiếp theo chứa 4 số nguyên a, b, c, d (1 ≤ a ≤ b ≤ n, 1 ≤ c ≤ d ≤ n).

Kết quả: Đưa ra thiết bị xuất chuẩn các câu trả lời xác định được cho mỗi truy vấn, mỗi câu trả lời trên một dòng.

Dữ liệu mẫu:

|   Input   |   Output  |
|-----------|-----------|
|abcbacaac  |    YES    |
|     2     |    NO     |
| 1 3 4 6   |           |
| 1 3 5 7   |           |

## 1.Abstraction

* Nhập vào q truy vấn. Với mỗi truy vấn thứ i nhập 4 số nguyên a, b, c, d. So sánh s**[**a:b**]** và s**[**c:d**]**. Nếu tương đương nhau thì xuất ra "YES" ngược lại xuất ra "NO". 2 xâu này chỉ tương đương khi tồn tại cách đổi chỗ các ký tự trong một xâu để được xâu kia.


## 2.Pattern recognization

* So sánh

## 3.Algorithm designed

Nếu tách xâu s thành các xâu s**[**a:b**]** và s**[**c:d**]** và đổi chỗ các ký tự cho đến khi 2 xâu này bằng nhau thì độ phức tạp quá lớn. Để giảm độ phức tạp ta đổi các ký tự trong bảng chữ cái thành dãy các bit nhị phân. Có 26 ký tự chữ cái thường sẽ cần 26 bit để biểu diễn. Ví dụ: a = 00...001 = 1, b = 00..010 = 2, với mỗi chữ cái tăng dần ta sẽ dịch bit 1 qua trái 1 bit.

* Bước 1: dùng dictionary trong python để ánh xạ ký tự chữ cái sang mã nhị phân.

* Bước 2: dùng 1 list với độ dài len(s) + 1, với s**[**0**]** = 0 để lưu giá trị của xâu s. List này sẽ tính sẵn các giá trị của xâu s, để giảm độ phức tạp xuống tối đa O(n). Ví dụ a = **[**1,5,2,3,2,5,2**]** thì list của chúng ta sẽ lưu các giá trị sau 
ls = **[**0,1,6,8,11,13,18,20**]**.Trong đó ls**[**i+1**]** = ls**[**i**]** + a**[**i**]** Khi đó nếu ta cần tính tổng từ i đến j thì ta chỉ cần lấy ls**[**j+1**]** - ls**[**i**]** sẽ ra tổng từ i -> j, độ phức tạp O(1).

* Bước 3: Tính tổng các ký tự trong xâu s**[**a:b**]** và tổng các ký tự trong xâu s**[**c:d**]** rồi sau đó so sánh với nhau. Nếu bằng nhau thì xuất ra "YES" ngược lại xuất ra "NO". 

## 4. Code
```php

from sys import stdin,stdout
s = stdin.readline().strip()
n = int(stdin.readline())
dic = {}
for i in range(97, 123):
    dic[chr(i)] = 2**(i-97)

dp = [0]*(len(s)+1)

for i in range(len(s)):
    dp[i+1] = dp[i] + dic[s[i]]

for i in range(n):
    a, b, c, d = map(int, stdin.readline().strip().split(' '))
    if dp[b] - dp[a-1] == dp[d] - dp[c-1]:
        print("YES")
    else:
        print("NO")
        
```
