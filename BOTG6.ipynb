### **3.Algorithm designed**

> *3.1 INPUT:*            
  n là số nguyên, 1 <= n <= 10^6      
  A là mảng lưu cái giá trị lãi thu được ở đoạn đường thứ i



> *3.2 KHỞI TẠO:*    
  max_so_far = âm vô hạn                
  max_ending_here = A[0]    
  start = end = temp = 0


> *3.3 Chạy vòng lặp i từ 0 đến n - 1 của mảng A :*


  a. max_ending_here = max_ending_here + a[i]

  b. if(max_so_far < max_ending_here)                                          
      cập nhật lại max_so_far = max_ending_here


  c. if(max_ending_here < 0)                                            
            max_ending_here = 0


> *3.4 RESULT:*

return max_so_far, start + 1, end + 1

 Độ phức tạp o(n)### **4.Programming:**
























```
n = int(input())
A = list (map(int, input().split()))

max_ending_here = A[0]
max_so_far = float ('-inf')
start = 0
end = 0 
temp = 0

for i in range(1,n):
  max_ending_here += A[i]

  if max_ending_here > max_so_far:
    max_so_far = max_ending_here
    end = i
    start = temp

  elif max_ending_here < 0:
    max_ending_here = 0
    temp = i + 1

print (start + 1, end + 1, max_so_far)
```
