### **Abstraction**


*   Tìm mảng con liên tiếp có tổng lớn nhất

### **Pattern recognization**


*   Quy hoạch động

### **Algorithm designed**

> *INPUT: * 
  n là số nguyên, 1 <= n <= 10^6      
  A là mảng lưu cái giá trị lãi thu được ở đoạn đường thứ i



> *KHỞI TẠO:*    
  max_so_far = âm vô hạn                
  max_ending_here = A[0]    
  start = end = temp = 0


> Chạy vòng lặp i từ 0 đến n - 1 của mảng A :


  1 max_ending_here = max_ending_here + a[i]

  2 if(max_so_far < max_ending_here)                                          
      cập nhật lại max_so_far = max_ending_here


  3 if(max_ending_here < 0)                                            
            max_ending_here = 0


> *RESULT:*

return max_so_far, start + 1, end + 1
























```
n = int(input())
A = list (map(int, input().split()))

max_ending_here = A[0]
max_so_far = float ('-inf')
start = 0
end = 0 
temp = 0

for i in range(1,n):
  max_ending_here += A[i]

  if max_ending_here > max_so_far:
    max_so_far = max_ending_here
    end = i
    start = temp

  elif max_ending_here < 0:
    max_ending_here = 0
    temp = i + 1

print (start + 1, end + 1, max_so_far)
```
